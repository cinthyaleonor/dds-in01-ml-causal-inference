# ¿Qué es (y qué no es) Inteligencia Artificial? 

---

Un modelo es considerado parte de la **Inteligencia Artificial (IA)** cuando cumple con ciertas características: 

### 1. **Automatización de la Toma de Decisiones**:
   - Un modelo de IA no sólo genera predicciones o clasificaciones, sino que también puede utilizar esos resultados para tomar decisiones de manera automatizada.
   - Ejemplo: Un modelo de IA en un sistema de recomendación predice tanto qué productos te gustarán como qué productos te podrían gustar y los sugiere de forma continua y personalizada, ajustándose a tu comportamiento.

### 2. **Aprendizaje Automático (Machine Learning)**:
   - Los modelos de IA generalmente se entrenan con datos y mejoran su rendimiento con el tiempo, sin necesidad de reprogramarse manualmente.
   - Ejemplo: Un modelo de aprendizaje supervisado que aprende de ejemplos etiquetados para predecir resultados futuros (como una red neuronal o un árbol de decisión).

### 3. **Capacidad de Generalización**:
   - Los modelos de IA buscan aprender patrones o representaciones generales a partir de los datos, permitiéndoles hacer predicciones o tomar decisiones en situaciones no vistas anteriormente.
   - Ejemplo: Un modelo de IA que aprende a identificar rostros en imágenes no solo reconoce imágenes específicas que ya ha visto, sino que puede identificar rostros en imágenes completamente nuevas.

### 4. **Adaptabilidad**:
   - A diferencia de los modelos tradicionales que están basados en fórmulas fijas, los modelos de IA pueden ajustarse a nuevas situaciones o condiciones, mejorando o adaptándose a partir de nueva información.
   - Ejemplo: Un sistema de IA que ajusta su comportamiento en función del cambio en las preferencias de los usuarios o en las condiciones de mercado.

### 5. **Procesamiento de Información No Estructurada**:
   - Muchas técnicas de IA, como las redes neuronales o los modelos de procesamiento del lenguaje natural (NLP), son capaces de trabajar con datos no estructurados, como texto, imágenes, audio y video, algo que los modelos tradicionales a menudo no pueden manejar fácilmente.
   - Ejemplo: Un modelo de IA que procesa imágenes para identificar objetos o clasifica correos electrónicos en función de su contenido textual.

### 6. **Capacidad de Manejar Complejidad**:
   - Los modelos de IA son capaces de manejar relaciones complejas y no lineales entre variables. Muchos algoritmos de IA, como las redes neuronales profundas (deep learning), pueden aprender patrones intrincados a partir de grandes volúmenes de datos.
   - Ejemplo: Un modelo de red neuronal que puede reconocer patrones en datos de alta dimensionalidad, como en el reconocimiento de voz o en la predicción de enfermedades.

### 7. **Interacción con el Entorno**:
   - Los modelos de IA pueden interactuar con su entorno y aprender de la retroalimentación que reciben, lo que les permite mejorar continuamente. Esto es especialmente relevante en IA de tipo **aprendizaje por refuerzo**.
   - Ejemplo: Un agente de IA en un videojuego que aprende a ganar partidas mejorando su estrategia con el tiempo.

### 8. **Comportamiento Inteligente**:
   - Un modelo de IA suele exhibir comportamientos que imitan la inteligencia humana, como reconocimiento de patrones, toma de decisiones, planificación, percepción y aprendizaje. 
   - Ejemplo: Un asistente virtual que comprende el lenguaje natural y responde a solicitudes como lo haría un humano.

### Diferencias con Modelos Tradicionales:
   - **Modelos tradicionales** (como la regresión lineal o logística básica) suelen tener una estructura fija y limitada en cuanto a su capacidad para aprender de datos más allá de los que fueron proporcionados inicialmente.
   - **Modelos de IA** tienden a aprender continuamente de datos, a manejar grandes volúmenes de información, a trabajar con datos no estructurados y a mejorar en función de la retroalimentación o interacción con el entorno.



Para que un modelo se considere IA, debe ir más allá de simples fórmulas o reglas predefinidas. Debe ser capaz de **aprender, adaptarse y tomar decisiones** de manera inteligente, muchas veces simulando el comportamiento humano o mejorando con el tiempo gracias a su interacción con datos nuevos o con su entorno.

# ¿Cuándo un modelo tradicional podría considerarse IA?

---
Un modelo, como la regresión logística, podría considerarse parte de un **sistema de IA** cuando:

1. **Se utiliza dentro de un proceso automatizado más amplio**:
   - Por ejemplo, si la regresión logística es parte de un sistema que continuamente se actualiza con nuevos datos y ajusta sus predicciones para mejorar su precisión o comportamiento.
   - Ejemplo: Un sistema de recomendación que utiliza regresión logística para predecir qué productos ofrecer, pero que aprende y se ajusta a partir de la interacción continua con los usuarios.

2. **Se integra con otras técnicas de Machine Learning o Deep Learning**:
   - Si la regresión es solo un componente de un sistema más amplio que utiliza, por ejemplo, técnicas de redes neuronales para el procesamiento de datos más complejos (como imágenes o texto), entonces el conjunto podría considerarse IA.
   - Ejemplo: Un asistente virtual que utiliza regresión logística para predecir la intención del usuario a partir del texto, pero está integrado en un sistema más amplio que entiende el lenguaje natural y aprende de las interacciones.

3. **El sistema es capaz de *aprender* y mejorar automáticamente**:
   - Si el modelo está en un bucle de retroalimentación, donde constantemente se ajusta y mejora con datos nuevos, entonces se acerca más a los sistemas de IA.
   - Ejemplo: Un modelo que recibe datos continuos sobre el comportamiento del usuario y ajusta sus parámetros para mejorar sus predicciones con el tiempo.
   
4. **Se combina con técnicas de procesamiento de datos no estructurados**:
   - Si el modelo se aplica a datos más complejos (como imágenes, texto o audio) y se integra con técnicas avanzadas de IA, entonces puede ser parte de un sistema de IA.
   - Ejemplo: Usar regresión logística para predecir la clasificación de imágenes dentro de un sistema que utiliza aprendizaje profundo para procesar esas imágenes.

### Diferencia clave:
Un **modelo tradicional** es más estático y no se adapta automáticamente. Por otro lado, un sistema de IA tiene capacidades más avanzadas que incluyen el **aprendizaje, la toma de decisiones automáticas y la adaptabilidad**. Cuando un modelo de regresión se utiliza dentro de un proceso que incorpora estas capacidades (por ejemplo, aprendizaje automático continuo), entonces podría considerarse parte de una **aplicación de IA**.

**La IA implica un proceso más amplio**, y los modelos tradicionales, como la regresión, pueden ser parte de ese proceso si se utilizan de manera que cumplan con los principios de **aprendizaje automático** y **automatización inteligente**.

## Definición de la Simulación

```plaintext
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

np.random.seed(42)
n = 10000

def generar_datos_iniciales(n):
    genero = np.random.choice(['M', 'F'], size=n)
    experiencia = np.random.randint(1, 15, size=n)
    puntuacion_entrevista = np.random.randint(1, 10, size=n)
    
    # Las mujeres tienden a tener mejores puntuaciones en la entrevista
    puntuacion_entrevista[genero == 'F'] += np.random.randint(0, 3, size=(genero == 'F').sum())
    puntuacion_entrevista = np.clip(puntuacion_entrevista, 1, 10)
    
    competencia = experiencia + puntuacion_entrevista
    
    # Simulamos un sesgo histórico en la selección
    resultado = []
    for g, c in zip(genero, competencia):
        if g == 'M':
            prob_seleccion = 0.3 + (c / 50)
        else:
            prob_seleccion = 0.2 + (c / 50)
        resultado.append(1 if np.random.random() < prob_seleccion else 0)
    
    return pd.DataFrame({
        'genero': genero,
        'experiencia': experiencia,
        'puntuacion_entrevista': puntuacion_entrevista,
        'competencia': competencia,
        'resultado': resultado
    })

def simular_refuerzo_sesgo(df, num_iteraciones=5):
    df_original = df.copy()
    modelo = LogisticRegression(random_state=42)
    
    # Listas para almacenar las tasas de selección
    tasas_hombres = []
    tasas_mujeres = []

    for i in range(num_iteraciones):
        X = df[['genero', 'experiencia', 'puntuacion_entrevista']]
        y = df['resultado']
        
        modelo.fit(X, y)
        
        # Generamos nuevos datos
        df_nuevo = generar_datos_iniciales(len(df) // 2)
        
        # Convertir género a numérico en el nuevo DataFrame
        df_nuevo['genero'] = df_nuevo['genero'].map({'M': 1, 'F': 0})
        
        X_nuevo = df_nuevo[['genero', 'experiencia', 'puntuacion_entrevista']]
        
        proba_nueva = modelo.predict_proba(X_nuevo)[:, 1]

        # Aplicar un umbral dinámico
        umbral_minimo = np.random.uniform(0.4, 0.6)  # Umbral variable entre 0.4 y 0.6
        seleccionados_nuevos = (proba_nueva > umbral_minimo).astype(int)

        # Asegurarse de que al menos el 10% de los candidatos sean seleccionados
        if seleccionados_nuevos.sum() < max(1, int(0.1 * len(seleccionados_nuevos))):
            seleccionados_nuevos[:int(0.1 * len(seleccionados_nuevos))] = 1

        df_nuevo['resultado'] = seleccionados_nuevos
        
        # Combinamos los datos originales con los nuevos
        df = pd.concat([df_original, df_nuevo]).reset_index(drop=True)
        
        # Calculamos y guardamos las tasas de selección por género
        tasa_hombres = df[(df['genero'] == 1) & (df['resultado'] == 1)].shape[0] / df[df['genero'] == 1].shape[0]
        tasa_mujeres = df[(df['genero'] == 0) & (df['resultado'] == 1)].shape[0] / df[df['genero'] == 0].shape[0]
        
        tasas_hombres.append(tasa_hombres)
        tasas_mujeres.append(tasa_mujeres)
        
        print(f"Iteración {i+1}: Tasa de selección - Hombres: {tasa_hombres:.2%}, Mujeres: {tasa_mujeres:.2%}")
    
    return df, modelo, tasas_hombres, tasas_mujeres
```

# Generamos datos iniciales y simulamos el refuerzo del sesgo

```plaintext
df_inicial = generar_datos_iniciales(n)
df_inicial['genero'] = df_inicial['genero'].map({'M': 1, 'F': 0})  # Conversión a numérico
df_sesgado, modelo_sesgado, tasas_hombres, tasas_mujeres = simular_refuerzo_sesgo(df_inicial)

# Preparamos los datos para la evaluación final
X = df_sesgado[['genero', 'experiencia', 'puntuacion_entrevista']]
y = df_sesgado['resultado']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Entrenamos un modelo sin sesgo (sin usar el género)
X_sin_sesgo = X[['experiencia', 'puntuacion_entrevista']]
X_train_sin_sesgo, X_test_sin_sesgo, _, _ = train_test_split(X_sin_sesgo, y, test_size=0.3, random_state=42)
modelo_sin_sesgo = LogisticRegression(random_state=42)
modelo_sin_sesgo.fit(X_train_sin_sesgo, y_train)

# Evaluamos ambos modelos
y_pred_sesgo = modelo_sesgado.predict(X_test)
y_pred_sin_sesgo = modelo_sin_sesgo.predict(X_test_sin_sesgo)

def evaluar_modelo(y_true, y_pred, genero, competencia, nombre_modelo):
    masculino = genero == 1
    femenino = genero == 0
    
    tasa_seleccion_m = y_pred[masculino].mean()
    tasa_seleccion_f = y_pred[femenino].mean()
    
    print(f"\nEvaluación del modelo {nombre_modelo}:")
    print(f"Tasa de selección para hombres: {tasa_seleccion_m:.2%}")
    print(f"Tasa de selección para mujeres: {tasa_seleccion_f:.2%}")
    print(f"Diferencia en tasas de selección: {tasa_seleccion_m - tasa_seleccion_f:.2%}")
    
    seleccion_injusta = ((y_pred == 1) & (genero == 1) & 
                         (competencia < competencia[genero == 0].mean())).sum()
    total_selecciones = (y_pred == 1).sum()
    porcentaje_injusto = seleccion_injusta / total_selecciones if total_selecciones > 0 else 0
    
    print(f"Porcentaje de selecciones donde se eligió a un hombre menos competente sobre mujeres más competentes: {porcentaje_injusto:.2%}")
    
    print("\nReporte de clasificación:")
    print(classification_report(y_true, y_pred))
    print("Matriz de confusión:")
    print(confusion_matrix(y_true, y_pred))

evaluar_modelo(y_test, y_pred_sesgo, X_test['genero'], df_sesgado.loc[X_test.index, 'competencia'], "con sesgo")
evaluar_modelo(y_test, y_pred_sin_sesgo, X_test['genero'], df_sesgado.loc[X_test.index, 'competencia'], "sin sesgo")
```

# Visualizaciones de tasas de selección


```plaintext

fig, ax = plt.subplots(figsize=(10, 6))
iteraciones = range(1, len(tasas_hombres) + 1)
ax.plot(iteraciones, tasas_hombres, label='Tasa de Selección - Hombres', marker='o', color='blue')
ax.plot(iteraciones, tasas_mujeres, label='Tasa de Selección - Mujeres', marker='o', color='deeppink')
ax.set_title('Refuerzo del Sesgo en la Selección por Iteración')
ax.set_xlabel('Iteraciones')
ax.set_ylabel('Tasa de Selección')
ax.set_ylim(0, 1)
ax.set_xticks(iteraciones)
ax.legend()
plt.grid()
plt.show()
```
